## Plotting with another dataset

```python
heart_disease = pd.read_csv("DATA/heart-disease.csv")
heart_disease.head()
```

```python
# Plotting graph for 50 years and above aged group
over_50 = heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = 'scatter',
             x = 'age',
            y = 'chol',
            c = 'target',
            ax = ax);
```

```python
# object oriented model
fig, ax  = plt.subplots(figsize = (10,6))

scatter = ax.scatter(over_50['age'],
                    over_50['chol'],
                    c = over_50['target'])
ax.set(title = 'Heart Disease and cholestrol levels',
      xlabel = 'Age',
      ylabel = 'Cholestrol',
      );
ax.legend(*scatter.legend_elements(), title = 'Target');
```

```python
#subplot of chol and thalach
fig, (ax1, ax2) = plt.subplots(2,1, figsize = (10,10), sharex = True)

ax1.scatter(over_50['age'],
           over_50['chol'],
            c= over_50['target'])

ax1.set(title = 'Age and cholesterol level',
        ylabel = 'Cholesterol')
        

ax1.legend(*scatter.legend_elements(), title ='target' )

ax2.scatter(over_50['age'],
           over_50['thalach'],
           c = over_50['target']);

ax2.set(title = 'Age and Max Heart Rate (thalach)',
        xlabel = 'Age',
        ylabel = 'Max Heart Rate')

ax2.legend(*scatter.legend_elements(), title ='target' )

fig.suptitle("Heart Disease Analysis", fontsize = 20, fontweight = "bold" );
```

### 3 Dimensional Contour plot.

```python
def f(x,y):
    return np.sin(x + y)

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
x, y = np.meshgrid(x,y)
z = f(x, y)
fig, ax = plt.subplots(subplot_kw = {'projection': '3d'})

ax.contour3D(x, y, z, 50, cmap = 'binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
# Create a meshgrid for x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate Z based on np.sin(x + y)
Z = np.sin(X + Y)

# Create a figure and a 3D axis
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Show the plot
plt.show()
```

### Creating a 3 dimensional animated plot from the above plot


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x and y values
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)

# Create the meshgrid for X and Y
x, y = np.meshgrid(x, y)

# Calculate z based on sine and cosine
z = np.sin(x) + np.cos(y)

# Create a figure and a 3D axis
fig, ax = plt.subplots(subplot_kw={'projection': '3d'})

# Hide the axes
ax.set_axis_off()

# Plot the 3D contour
ax.contour3D(x, y, z, 100, cmap='viridis')

# Set panel color to white
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))

# Make grid lines invisible
ax.xaxis._axinfo['grid']['color'] = (1, 1, 1, 0)
ax.yaxis._axinfo['grid']['color'] = (1, 1, 1, 0)
ax.zaxis._axinfo['grid']['color'] = (1, 1, 1, 0)

# Adjust the point of view
ax.view_init(40, 90)

# Show the plot
plt.show()
```

**Note:** For an animated video we need multiple images of the same graph from different point of view and save it in a folder, the following function does the same.

```python
import os

# Define the x and y values
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
x, y = np.meshgrid(x, y)
z = np.sin(x) + np.cos(y)

# Create output directory if it doesn't exist
output_dir = '3d'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Generate plots from different angles
for angle in range(0, 360, 2):
    plt.figure(figsize=(16, 9))
    ax = plt.axes(projection='3d')
    
    # Hide the axes
    ax.set_axis_off()
    
    # 3D contour plot
    ax.contour3D(x, y, z, 200, cmap='viridis')
    
    # Set panel colors to white
    ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    
    # Make grid lines invisible
    ax.xaxis._axinfo['grid']['color'] = (1, 1, 1, 0)
    ax.yaxis._axinfo['grid']['color'] = (1, 1, 1, 0)
    ax.zaxis._axinfo['grid']['color'] = (1, 1, 1, 0)
    
    # Adjust view based on the angle
    ax.view_init(40, angle)
    
    # Save the figure with a filename based on the angle
    filename = os.path.join(output_dir, f'3d_vis_{angle}.png')
    plt.savefig(filename, dpi=75)
    plt.close()  # Close the figure to free up memory
```

```python
# Create the frames
from PIL import Image

# Directory where images are stored
directory = '3d'

# Create the frames
frames = []
png_count = 180
for i in range(png_count):
    seq = str(i * 2)
    file_name = os.path.join(directory, f'3d_vis_{seq}.png')
    if os.path.exists(file_name):  # Ensure the file exists
        new_frame = Image.open(file_name)
        frames.append(new_frame)
    else:
        print(f"File {file_name} not found.")

# Save into a GIF file that loops forever
if frames:
    frames[0].save('3d_vis.gif', format='GIF',
                   append_images=frames[1:],
                   save_all=True,
                   duration=40, loop=0)
else:
    print("No frames to save.")
```
